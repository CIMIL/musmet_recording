```python
try:
    with socket.create_connection(("192.168.0.106", 22346), timeout=3) as s:
        s.sendall(b"start\n")
        # print(f"Sent {len('start\n')} bytes")
        response = s.recv(1024)  # Wait for acknowledgment
        print("Server response:", response.decode())
except ConnectionRefusedError:
    print("Target machine refused connection - check if server is running")
except TimeoutError:
    print("Connection timed out - check firewall/network settings")
except Exception as e:
    print(f"Error: {str(e)}")
```
